```python
import pandas as pd
import datetime
from sklearn.preprocessing import LabelEncoder
import numpy as np
le = LabelEncoder()
df = pd.read_csv(r'C:\Users\nCalo\Documents\Automifai\Research\Coding_Lessons\Git\NYC_apartment_price_project\Data\2_cleaned_NYC_property_sales.csv')
```

```python
df.head(15)
```

```python
df = pd.get_dummies(df, columns=['NEIGHBORHOOD', 'BUILDING CLASS CATEGORY', 'TAX CLASS AT TIME OF SALE'])
```

```python
df_neighborhood_cols = [col for col in df.columns if 'NEIGHBORHOOD' in col]
print(df_neighborhood_cols)
```

```python
df['current_year'] = datetime.datetime.now().year
df['YEAR BUILT'].astype(int)
```

```python
df['AGE OF BUILDING'] = df['current_year']-df['YEAR BUILT']
df.head(10)
```

```python
bins = [0,3,10,20,30,50,75,100,150,1000]
labels = [1,2,3,4,5,6,7,8,9]
df['AGE OF BUILDING'] = pd.cut(df['AGE OF BUILDING'], bins=bins, labels=labels, right=True)
```

```python
df.head(10)
```

```python
df['AGE OF BUILDING']=df['AGE OF BUILDING'].astype('object') #change age of building type to an object in order to allow one hot encoding
df['AGE OF BUILDING']
```

```python
df = pd.get_dummies(df, columns=['AGE OF BUILDING']) #one hot encode building age
```

```python
df.head(10)
```

```python
df = df.drop(columns=['#','BOROUGH','BLOCK', 'YEAR BUILT', 'current_year', 'TOTAL UNITS', 'LAND SQUARE FEET', 'BUILDING CLASS AT TIME OF SALE'])
#drop columns we don't need
df.head(10)
```

```python
df_SALEPRICE = df.pop('SALE PRICE') #move sale price to the last column
df['SALE PRICE']=df_SALEPRICE
df.head(10)
```

```python
df = df.rename(columns=lambda x: x.strip()) # cleaning excess text for columns
df.columns = [col.replace('NEIGHBORHOOD_', '') for col in df.columns]
df.columns = [col.replace('BUILDING CLASS CATEGORY_', '') for col in df.columns]
df.columns = [col.replace('TAX CLASS AT TIME OF SALE_', '') for col in df.columns]
```

```python
df.columns.values
```

```python
df['SALE PRICE'].max()
```

```python
df['SALE PRICE'] = df['SALE PRICE'][np.abs(df['SALE PRICE']-df['SALE PRICE'].mean()) <= (.3*df['SALE PRICE'].std())] #get rid if outliers
```

```python
df['SALE PRICE'].max()
```

```python
df['SALE PRICE'].replace('', np.nan, inplace=True)
df.dropna(subset=['SALE PRICE'], inplace=True)
```
